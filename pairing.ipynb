```python
#creates a new dataframe/csv for all pairs of protein combinations within a group
```

```python
import pandas as pd
import numpy as np
```

```python
proteins = pd.read_csv("filteredtrain.csv")
```

```python
def pairing(original_df, starting_gn, final_gn):
    pairs = pd.DataFrame(columns = ['sequence_one', 'sequence_two', 'dtm', 'size', 'group_number'])
    
    gn = 0
    
#     starting_gn = 112
    
#     final_gn = 336

    counter = 0
    
    for gn in range(starting_gn, final_gn + 1):
        gn_section = (proteins.loc[proteins.group_number == gn]).copy()
        gn_section = gn_section.reset_index()
        section_size = gn_section.shape[0]
        
        tm_one = 0
        
        
        
        #print('current group number: ', gn,'\n')
        
        if (gn%10 == 0):
            print('Entries generated: ', counter,'\n')
        
        
        for i in range(0, section_size):
            if i == 0:
                size = gn_section.loc[i, 'Size']
                group_number = gn_section.loc[i, 'group_number']
                
                
            tm_one = gn_section.loc[i, 'tm']
            sequence_one = gn_section.loc[i, 'protein_sequence']
            
            
            j = i + 1
            
            while j < section_size:
                sequence_two = gn_section.loc[j, 'protein_sequence']
                tm_two = gn_section.loc[j, 'tm']
                
                new_entry = pd.DataFrame({'sequence_one': sequence_one, 'sequence_two': sequence_two, 'dtm': tm_one - tm_two, 
                                      'size': size, 'group_number': group_number}, index=[0])
                
                
                pairs = pd.concat([pairs, new_entry], ignore_index=True)
                
                #the following entry is the same as the one before but reversed to ensure we get both types to not bias our data
                
                new_entry = pd.DataFrame({'sequence_one': sequence_two, 'sequence_two': sequence_one, 'dtm': tm_two - tm_one, 
                                      'size': size, 'group_number': group_number}, index=[0])
                
                pairs = pd.concat([pairs, new_entry], ignore_index=True)
                
                j += 1
                
                counter +=2
                
    return pairs
                
```

```python
starting_gn = 209 #size 112

final_gn = 867 #size 336
    
    
pairos = pairing(proteins, starting_gn, final_gn)
```
